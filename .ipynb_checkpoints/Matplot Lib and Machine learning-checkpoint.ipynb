<font color ="pink" size =18>Functions in Lambda</font>


```python
#lambda functions
x =lambda x,y:x+y
g =lambda x:x**2 if x>100  else x**3
print(x(123,789))
g(99)
```

```python
#map
J =[100,34,3]
list(map(g,J))
```

```python
#filter
B =lambda x: x<23
u =[12,34,56,1,10]
list(filter(B,u))
```

<font color ="pink" size =18>Graphing in Matplotlib</font>


```python
import pandas as pd
import matplotlib.pyplot as plt
import os
os.listdir()
df =pd.read_excel('Test.xlsx')
df.fillna(0)
H  = df.groupby('Expense Description')
New = H[['Amount', 'VAT', 'Total']].agg('sum')
New.head()
```

```python
#plot 1
%matplotlib inline
plt.figure(figsize =(10,10))
plt.bar(New.index,New.Amount)
plt.title("Expenses graph")
plt.xlabel("expenses")
plt.xticks(rotation =60)
plt.ylabel("Amunt in KES")
#plt.legend()
plt.show()

#plot2
plt.figure(figsize =(10,10))
plt.barh(New.index,New.Amount)
plt.title("Expenses graph")
plt.xlabel("expenses")
#plt.xticks(rotation =60)
plt.ylabel("Amunt in KES")
#plt.legend()
plt.show()

```

```python
#plot3
plt.plot(New.Total,'g-*')
plt.bar(New.index,New.Amount,color ="yellow")
plt.xticks(rotation =60)
plt.show()

#plot 4
plt.figure(figsize=(5,5))
plt.pie(New.Total,labels =New.index)
plt.legend()
plt.show()
```

<font color ="pink" size =18>Simple Machine learning</font>


```python
import numpy as np
from sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as MAE
```

```python
import random
D =np.random.randint(0,1000,(100,3))
df =pd.DataFrame(D,columns=["Length","Width","Height"])
df["Price"] =df.sum(axis=1)*random.randint(60,65)*100
x =df[["Length","Width","Height"]]
y =df.Price

x_train,x_test,y_train,y_test =train_test_split(x,y, test_size =0.8)
```

```python
model = LR()
model.fit(x_train,y_train)
#test the model
pred =model.predict(x_test)
re =MAE(pred,y_test)
print(re)
```

```python
B =pd.DataFrame()
B['Actual'] = y_test
B['Predicted'] =pred
B.head()
```

```python
#You have new unseen  data that you want to predict
G =[23,41,99]
G =np.array([G])
model.predict(G)
```
